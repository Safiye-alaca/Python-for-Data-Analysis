#### Using HDF5 Format

```python
frame = pd.DataFrame({"a": np.random.standard_normal(100)})
store = pd.HDFStore("C:/Users/safye/Downloads/pydata-book-3rd-edition/pydata-book-3rd-edition/examples/mydata.h5")
```

```python
store["obj1"] = frame
```

```python
store["obj1_col"] = frame["a"]
```

```python
store
```

```python
type(store)
```

```python
store["obj1"]
```

```python
store["obj1_col"]
```

```python
store.put("obj2", frame, format="table")
```

```python
store.select("obj2", where=["index >= 10 and index <=15"])
```

### 6.4 Veritabanları ile Etkileşim

```python
import sqlite3
```

```python
query = """CREATE TABLE test(a VARCHAR(20), b VARCHAR(20), c REAL, d INTEGER);"""
```

```python
can = sqlite3.connect("mydata.sqlite")
```

```python
can.execute(query)
```

```python
type(can)
```

```python
data = [("Atlanta", "Georgia", 1.25, 6),("Tallahassee", "Florida", 2.6, 3),
        ("Sacramento", "California", 1.7, 5)]
```

```python
stmt = "INSERT INTO test VALUES(?, ?, ?, ?)"
```

```python
can.executemany(stmt, data)
```

```python
can.commit()
```

```python
cursor = can.execute("SELECT * FROM test")
```

```python
rows = cursor.fetchall()
```

```python
rows
```

```python
cursor.description
```

```python
import sqlalchemy as sqla
db = sqla.create_engine("sqlite:///mydata.sqlite")
```

```python
pd.read_sql("SELECT * FROM test", db)
```
